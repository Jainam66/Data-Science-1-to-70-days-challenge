```python
#scatter Plot
#gives the relationship between the columns
ages=[30,38,45,29,50,47]
salary=[3000,4000,5000,6000,2934,1234]
plt.xlabel("Age of Person")
plt.ylabel("Salary of the Person")
plt.title("Salary Distribution of the Employee")
plt.scatter(ages,salary,color="green")
```

```python
#Line Visualization
x=[-5,-2,-1,1,3,4,5,10]
y=[35,34,50,60,20,50,90,80]
plt.figure(figsize=(6,4)) #dimension of the figure
plt.suptitle("Learning Visualization",size=25)
plt.xlim(-20,20)
plt.ylim(10,100)
plt.plot(x,y)
```

```python
 #plot the graph of no (-100 to 100) and its square using the range function
 x=list(range (-100,100))
 y=[i**2 for i in x]
 plt.plot(x,y)
```

```python
from math import exp
x=list(range(0,30))
y=[exp(i) for i in x]
plt.plot(x,y)
```

```python
df1=pd.read_csv("/content/mtcars.csv")
df1
```

```python
df1.shape
```

```python
#find no of cars having certain no of cylinders and certain no of gairs
pd.crosstab(df1["cyl"],df1["gear"])
#data schows that 4 cyl 3 Gear has 1 car and etc...
```

```python
#bargraph
pd.crosstab(df1["cyl"],df1["gear"]).plot(kind='bar')
```

```python
df1["gear"].value_counts()
```

```python
df1.gear.value_counts().plot(kind='pie')
```

```python
#scatterplot between 2 columns(mpg and Qsec)
plt.scatter(df1.mpg,df1.qsec,color="green")
plt.xlabel=("mpg")
plt.ylabel=("qsec")
plt.title=("mpg vs qsec plot")
```

```python
plt.hist(df1.mpg,edgecolor="green" )
```

```python
plt.hist(df1.mpg,facecolor="yellow",edgecolor="green",bins=6)
#6 categories are created
```

```python
#boxplot___OUTLIERS_Detector
plt.boxplot(df1.mpg)
plt.show()
```

```python
plt.boxplot(df1.mpg,vert=False,notch=True )
plt.show()
```
